```python
import matplotlib.pyplot as plt

import csv
import numpy as np

# 记得改路径

# 普通的
with open("D:/group/reinforcement/data/local/20240512/LunarLanderExperiment_mlp/run_0/progress.csv", 'r') as f:
    reader = csv.DictReader(f)
    column_mlp = [row['ReturnAverage'] for row in reader]
    
    
# 对称的    
with open("D:/group/reinforcement/data/local/20240512/LunarLanderExperiment_equivariant/run_0/progress.csv",'r') as f:
    reader = csv.DictReader(f)
    column_equivariant = [row['ReturnAverage'] for row in reader]
    
    
    
for i in range(len(column_mlp)):
    if column_mlp[i] == 'nan':
        column_mlp[i] = column_mlp[i - 1]
        
for i in range(len(column_equivariant)):
    if column_equivariant[i] == 'nan':
        column_equivariant[i] = column_equivariant[i - 1]
        
column_mlp = [float(i) for i in column_mlp]
column_equivariant = [float(i) for i in column_equivariant]


column_mlp = np.convolve(column_mlp, [0.3, 0.4, 0.3], mode='same') # 平滑化处理
column_equivariant = np.convolve(column_equivariant, [0.3, 0.4, 0.3], mode='same')
```

# 橙色是普通的，蓝色是对称的

```python
plt.plot(column_mlp)
plt.plot(column_equivariant)

plt.show()
```
